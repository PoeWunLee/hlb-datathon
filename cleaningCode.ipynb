```python
import pandas as pd

df = pd.read_csv("mortgage_dataset.csv")
```

```python
df.drop(columns=["Application_Number","Land_Area_Unit","Property_Address","Property_Description","Build_Up_Area_Unit"], inplace=True)
```

```python
df
```

```python
def getUnique(thisDF):
    
    for columns in df.columns:
        print(thisDF[columns].unique())

getUnique(df)
```

```python
#replace empty string with NaN, and then drop NaN rows
df["Free_Lease_Hold_Ind"].loc[df["Free_Lease_Hold_Ind"]==" "] = float("NaN")
df.dropna(inplace=True)
```

```python
#reformatting strings
df["Property_State"] = df["Property_State"].str.lower()
df["City_Nm"] = df["City_Nm"].str.lower()
df["Disbursement_Manner"] = df["Disbursement_Manner"].str.strip()

#property name type cleaning
replacePropertyDict = {
    "Condo": "Condominium",
    "Shophouse/Shop office/Shop lot/Retail lot": "Shop lot",
    "Shop/Office": "Office lot",
    'Town House' : "Townhouse"
}
df = df.copy()
def replaceApartmentNames(replaceDict):
    for key,value in replaceDict.items():
        df["Property_Type"].loc[df["Property_Type"]==key] = value

replaceApartmentNames(replacePropertyDict)
df = df.copy()
df["Residential_Type"] = df["Residential_Type"].str.slice(start=3)
df
```

```python
def getCategorical(df,cols):
    df= df.copy()
    for thiscol in cols:
        df[thiscol] = pd.Categorical(df[thiscol])
        print(df[thiscol].cat.categories)
        df[thiscol] = df[thiscol].cat.codes
    return df

categoryCols = ["Disbursement_Manner","Free_Lease_Hold_Ind", "Status_Of_Completion", "Title_Type", "Property_State", "Property_Type", "Residential_Type", "Landed_Type","City_Nm"]

df_cat = getCategorical(df,categoryCols)
    

```
