```python
def filterDF(restype,proptype,land,df):
    df = df.copy()
    iterDict = {
        "Residential_Type": restype,
        "Property_Type": proptype,
        "Landed_Type": land
    }
    
    for key,value in iterDict.items():
        print(key,value)
        df = df.loc[df[key] == value]
        
    return df

filtDF = filterDF("Residential", "Condominium", "Non-Landed",df)

labels = ["SnP > OWP", "SnP < OWP"]
#labels[0], labels[1] = "SnP > OWP", "SnP < OWP"
values = [filtDF.loc[filtDF["is_SnP_higher"]==True]["is_SnP_higher"].count(), filtDF.loc[filtDF["is_SnP_higher"]==False]["is_SnP_higher"].count()]

fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent',
                             insidetextorientation='radial', hole=.7
                            )])
fig.update_layout(template="plotly_dark", paper_bgcolor="#303030")
```
